```python
from langchain.evaluation import load_evaluator
from langchain_openai import ChatOpenAI

evaluator = load_evaluator("labeled_score_string", llm=ChatOpenAI(model="gpt-3.5-turbo"))
```

```python
# Correct
eval_result = evaluator.evaluate_strings(
    prediction="The admission requirements for the M.Sc. in Applied Computer Science include having a Bachelor's degree in computer science or a related field, with a minimum GPA as specified by the university.",
    reference="Applicants need a Bachelor's degree in computer science or a related field, with a minimum GPA as specified by the university.",
    input="What are the admission requirements for the M.Sc. in Applied Computer Science?",
)
print(eval_result)
```
